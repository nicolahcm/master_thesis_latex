```python
from fbm import FBM
import matplotlib.pyplot as plt

%matplotlib
```

```python
f = FBM(n=1024, hurst=0.7, length=1, method='daviesharte')
```

```python
f.times()
```

```python
fbm_sample = f.fbm()
```

```python
fbm_sample
```

```python
fig, ax = plt.subplots()
```

```python
ax.plot(f.times(),fbm_sample, label="random fBm")
```

```python
ax.legend()
ax.set_xlabel('Time')
ax.set_title('fBm with H=0.9')
```
